```python
import numpy as np
import plotly.express as px
import pandas as pd
import spacy
import re
from func_spacy_actor import spacy_actor, unify_to_train_test
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB, GaussianNB
from sklearn.pipeline import make_pipeline
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
artist_filesp=['aero_disco.p','beatles_disco.p','blacksabb_disco.p','cash_disco.p','elvis_disco.p','frank_disco.p','led_disco.p','pink_disco.p','rolling_disco.p','who_disco.p']
X_trainall, X_testall, y_trainall, y_testall = unify_to_train_test(artist_filesp)
X_trainall=spacy_actor(X_trainall)
X_testall=spacy_actor(X_testall)
```

```python
cv = CountVectorizer()
tf = TfidfTransformer()
nb = MultinomialNB(alpha=0.01)
cv.fit(X_trainall)
bow_corpus = cv.transform(X_trainall)
tf.fit(bow_corpus)
features_train = tf.transform(bow_corpus)
features_test = tf.transform(cv.transform(X_testall))
```

```python
features = np.concatenate((features_train.todense(),features_test.todense()), axis=0)
labels = np.concatenate((y_trainall,y_testall), axis=0)
```

```python
np.unique(labels)
```

```python
mod = PCA(n_components=2)
principal_components = mod.fit_transform(features)
df_features = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
df_labels = pd.DataFrame(data=labels, columns=['label'])
df_full = pd.concat([df_features, df_labels], axis=1)
df_full['label'] = df_full['label'].astype(str)
plt.figure(figsize=(10,10))
sns.scatterplot(x='PC1',
                    y='PC2',
                    hue=labels,                
                    data=df_full,
                    alpha=.7).set_title('PCA decomposition');
```

```python
mod = PCA(n_components=3)
principal_components = mod.fit_transform(features)
df_features = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2', 'PC3'])
df_labels = pd.DataFrame(data=labels, columns=['label'])
df_full = pd.concat([df_features, df_labels], axis=1)
df_full['label'] = df_full['label'].astype(str)
```

```python
fig = px.scatter_3d(df_full, x='PC1', y='PC2', z='PC3', color='label')
fig.show()
```
